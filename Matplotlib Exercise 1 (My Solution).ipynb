## Exercise 3
Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]

```python
#Do not run this code please ! 
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax1.plot(x,z)
ax1.set_xlabel('x')
ax1.set_ylabel('z')
ax1.set_xlabel('x')
ax1.set_ylabel('z')
ax1.set_xlim(0,100)
ax1.set_ylim(0,10000)

ax2 = fig.add_axes([0.2,0.5,0.4,0.4])
ax2.plot(x,y)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('zoom')
ax2.set_xlim(20,22)
ax2.set_ylim(30,50);
```
